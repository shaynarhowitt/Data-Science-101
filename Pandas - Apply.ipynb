```python
#Student alcohol consumption exercise

import pandas as pd
data = pd.read_csv('student-mat.csv')
data.head()
```

```python
#slice dataframe from 'school' until 'guardian'

data = data.loc[ : , 'school':'guardian']
data.head()
```

```python
def capitalize(x):
    return x.capitalize()
data['Fjob'] = data['Fjob'].apply(capitalize)
data['Mjob'] = data['Mjob'].apply(capitalize)
data.head()
```

```python
#create function called majority that checks if age > 17
#return a boolean value to a new column called legal_drinker 

def majority(x):
    return x >= 17

data['legal_drinker'] = data['age'].apply(majority)
data.head()
```

```python
#multiply every number of the dataset by 10

def times10(x):
    return x * 10

data.applymap(times10).head()
```
