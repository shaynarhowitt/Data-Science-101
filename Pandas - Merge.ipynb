```python
# Fictitious Names Exercise

raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}

data1 = pd.DataFrame(raw_data_1)
data2 = pd.DataFrame(raw_data_2)
data3 = pd.DataFrame(raw_data_3)
```

```python
# Join along rows

all_data = pd.concat([data1, data2])
all_data
```

```python
all_data_col = pd.concat([data1, data2], axis = 1)
all_data_col
```

```python
data3
```

```python
#merge all_data and data3 along subject_id value

pd.merge(all_data, data3, on='subject_id')
```

```python
#merge only data that has the same subject_id on both data1 and data2

pd.merge(data1, data2, on="subject_id", how="inner")
```

```python
#merge with matching records from both sides when available

pd.merge(data1, data2, on='subject_id', how='outer')
```

```python
#Housing Market Exercise

np.random.normal()
s1 = np.random.randint(low = 1, high = 4, size=100)
s2 = np.random.randint(low = 1, high = 3, size=100)
s3 = np.random.randint(low = 10000, high = 30000, size=100)
```

```python
df1 = pd.DataFrame(s1)
df2 = pd.DataFrame(s2)
df3 = pd.DataFrame(s3)
```
