```python
#Alcohol consumption exercise

import pandas as pd
```

```python
url = ('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv')
drinks = pd.read_csv(url, sep = ',')
drinks.head()
```

```python
#Which continent drinks more on average?

drinks['total'] = drinks['beer_servings'] + drinks['spirit_servings'] + drinks['wine_servings']
by_cont = drinks.groupby('continent')
by_cont['total'].mean().sort_values(ascending = False)
```

```python
by_cont['wine_servings'].sum()
```

```python
by_cont.median()
```

```python
by_cont['spirit_servings']
```

```python
# Occupation exercise

url1 = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(url1, sep = '|')
users.head()
```

```python
occ_group = users.groupby('occupation')
occ_group['age'].mean()
```

```python
#Male to female ratio per occupation

grouped_users = users[['occupation', 'gender']].groupby('occupation')
grouped_users['gender'].value_counts()
```

```python
#calculate mean age for each combination of occupation and gender

users.groupby(['occupation', 'gender']).age.mean()
```
