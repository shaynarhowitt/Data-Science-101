```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv'
baby_names = pd.read_csv(url)
baby_names.head(10)
```

```python
names = baby_names.groupby("Name").sum()
print(names.shape)
names.sort_values("Count", ascending = 0).head()
```

```python
#name with most occurences

names[names.Count == names.Count.max()]
```

```python
#How many names have the least occurences?

len(names[names.Count == names.Count.min()])
```

```python
#wind statistics exercise

url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data'
```

```python
data = pd.read_table(url, sep = "\s+", parse_dates = [[0,1,2]]) 
data.head()
```

```python
#correct that years are from 2061

import datetime 

def fix_century(x):
  year = x.year - 100 if x.year > 1989 else x.year
  return datetime.date(year, x.month, x.day)

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_century)
data.head()
```

```python
data.set_index('Yr_Mo_Dy').head()
```

```python
data.index = pd.to_datetime(data.index)
```

```python
#find number of NaN instances in a given column

len(data.index) - data['RPT'].count()
```

```python
#find number of NaN instances across all columns

data.isnull().sum()
```

```python
#count total number of non-missing values

data.count().sum()
```

```python
data.mean()
```

```python
#Calculate the mean windspeeds of the windspeeds over all the locations and all the times.

data.mean().mean()
```

```python
#calculate the min, max and mean windspeeds and standard deviations of the windspeeds at
#each location over all the days

loc_stats = pd.DataFrame()

loc_stats['min'] = data.min() # min
loc_stats['max'] = data.max() # max 
loc_stats['mean'] = data.mean() # mean
loc_stats['std'] = data.std() # standard deviations

loc_stats
```
